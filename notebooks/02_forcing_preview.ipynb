```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = '../data/ERA5/reanalysis-era5-single-levels-timeseries-sfc6369p5v8.nc'
ds = xr.open_dataset(path)
ds
```

```python
rho_air = 1.225
Cd = 1.3e-3
alphaT = 15.0
alpha_sw = 1e-6
alpha_lw = 1e-6

u10 = ds['u10'].values.astype(float)
v10 = ds['v10'].values.astype(float)
t2m = ds['t2m'].values.astype(float)
sst = ds['sst'].values.astype(float)
ssrd = ds['ssrd'].values.astype(float)
strd = ds['strd'].values.astype(float)
time = pd.to_datetime(ds['valid_time'].values)

U = np.sqrt(u10**2 + v10**2)
tau_x = rho_air * Cd * U * u10
tau_y = rho_air * Cd * U * v10
QT = alphaT * (t2m - sst) + alpha_sw * ssrd + alpha_lw * strd

df = pd.DataFrame({
    'time': time,
    'u10': u10, 'v10': v10, 't2m': t2m, 'sst': sst, 'ssrd': ssrd, 'strd': strd,
    'tau_x': tau_x, 'tau_y': tau_y, 'QT': QT,
})
df.head()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)
axes[0].plot(df['time'], df['u10'], label='u10')
axes[0].plot(df['time'], df['v10'], label='v10')
axes[0].legend()
axes[0].set_title('10 m winds')

axes[1].plot(df['time'], df['tau_x'], label='tau_x')
axes[1].plot(df['time'], df['tau_y'], label='tau_y')
axes[1].legend()
axes[1].set_title('Wind stress targets')

axes[2].plot(df['time'], df['QT'], label='QT')
axes[2].legend()
axes[2].set_title('Heat-flux proxy target')
plt.tight_layout()
```
